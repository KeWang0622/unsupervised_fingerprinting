```python
% matplotlib notebook
import sigpy.plot as pl
import numpy as np
import torch
import torch.backends.cudnn as cudnn
import torch.nn as nn
import torch.nn.functional as F
from torch import optim
import scipy.io
from torch.autograd import Variable
import UFNet
import os
import resnet
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
%load_ext autoreload
%autoreload 2
```

```python
#Knee data folders: 
folder_knee = "/mikRAID/frank/data/cube_knees/train_ksp_slices/"
files_knee = os.listdir(folder_knee)
# len(files_knee)/
files_knee_valid = list([])
for i in files_knee:
        if os.path.splitext(i)[1] == '.npy':
            files_knee_valid.append(i)
```

```python
def dir2dic(dir_list):
    n = len(dir_list)
    output = list([])
    for i in range(n):
        dir_n = dir_list[i]
        dk = list([dir_n])
        dk.extend(dir_n[:-4].split("_"))
        output.append(dk)
    return output
# files_knee_valid[0][:-4].split("_")
```

```python
# Construct dir dictionary and permuta the number
dir_dic = dir2dic(files_knee_valid)
n_elements = len(dir_dic)
dic_num = np.random.permutation(n_elements)
```

```python
def index2torch(index_dic,dir_dic,folder):
    n_index = index_dic.shape[0]
    for jj in range(n_index):
        if jj == 0:
            outp = np.load(folder+dir_dic[index_dic[0]][0])[None,:,:,:]
        else:
            outp = np.concatenate((outp,np.load(folder+dir_dic[index_dic[jj]][0])[None,:,:,:]))
    return torch.tensor(np.concatenate((outp.real,outp.imag),1))
```

```python
# Hyper parameter tuning
EPOCHS = 50
BATCH_SIZE = 8
LR = 0.001
```

```python
n_features = 3
# Define the bank
B_tensor_cuda = torch.zeros((n_elements,n_features)).cuda()
```

```python
# Define the network
ksnet = resnet.resnet50(num_classes=n_features).cuda()
n = 160
steps = int(n_elements/n)
with torch.no_grad(): 
    for j in range(steps):
        ip = np.array(range(j*n,(j+1)*n))
        inp = index2torch(ip,dir_dic,folder_knee).cuda()
        out = ksnet(inp)
#         break
        B_tensor_cuda[j*n:(j+1)*n,:] = out
        print(j)
```

```python
pl.LinePlot(B_tensor_cuda.detach().cpu().numpy())
```

```python
optimizer = torch.optim.Adam(ksnet.parameters(),
                          lr=0.0001)
criterion = torch.nn.CrossEntropyLoss()
tau = 1
loss_all = list([])
index_dic = np.random.permutation(n_elements)
batchsize = 16
n_step = int(n_elements/batchsize)

num = np.floor(n_elements/batchsize).astype(int)

for epoch in range(EPOCHS):
    index_dic = np.random.permutation(n_elements)
    ksnet.train()
    epoch_loss = 0
    optimizer.zero_grad()
    print("Starting Epoch: %d" %(epoch+1))
    torch.save(ksnet.state_dict(),
                           'CP_unsuper{}.pth'.format(epoch + 1))
    np.save("kspace_unsupervised_resnet.npy",B_tensor_cuda.detach().cpu().numpy())
    for index in range(n_step):
        ipd = index_dic[index*batchsize:(index+1)*batchsize]
#         break
        input_dic = index2torch(ipd,dir_dic,folder_knee).cuda()
        output_dic = ksnet(input_dic)
#         break
        ## error: Trying to backward through the graph a second time, but the buffers have already been freed. Specify retain_graph=True when calling backward the first time.
#             print(output_dic)
#             B_tensor_cuda.requires_grad = False
        output_dic1 = torch.mm(output_dic,B_tensor_cuda.detach().float().transpose(0,1))/tau
#             print(output_dic.requires_grad)
#             print(output_dic.shape)
#         ipd = ipd % 22031
        loss = criterion(output_dic1,torch.LongTensor(ipd).cuda())
        if index % 5 == 0:
            print(loss.item())
            loss_all.append(loss.item())
        epoch_loss = epoch_loss+ loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        B_tensor_cuda[ipd,:] = output_dic
        
        
```
